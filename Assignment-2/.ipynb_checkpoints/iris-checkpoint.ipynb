
# The Iris Dataset

This data sets consists of 3 different types of irises'
(Setosa, Versicolour, and Virginica) petal and sepal
length, stored in a 150x4 numpy.ndarray

The rows being the samples and the columns being:
Sepal Length, Sepal Width, Petal Length	and Petal Width.
(Sepal: 萼片, Petal: 花瓣, length and width is in centimeters）

The below plot uses the first two features.
See `here <https://en.wikipedia.org/wiki/Iris_flower_data_set>`_ for more
information on this dataset.



```python
import matplotlib.pyplot as plt
from sklearn import datasets

# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :2]  # we only take the first two features.
y = iris.target

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5

plt.figure(2, figsize=(8, 6))
plt.clf()

# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1,
            edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())

plt.show()
```

### 1. What is the min, max, mean for each type of feature (Sepal Length, Sepal Width, Petal Length and Petal Width)?  Which type do you think a flower with Seqal (length=4.7, width=3.2), Pedal (length=1.6, width=0.2) belong to?

```python
print('min of Sepal Length: ', X[:,0].min())
print('max of Sepal Length: ', X[:,0].max())
print('mean of Sepal Length: ', round(X[:,0].mean(),2))
print('---------------')
print('min of Sepal Width: ', X[:,1].min())
print('max of Sepal Width: ', X[:,1].max())
print('mean of Sepal Width: ', round(X[:,0].mean()))
print('---------------')
print('min of Petal Length: ', iris.data[:,2].min())
print('max of Petal Length: ', iris.data[:,2].max())
print('mean of Petal Length: ', round(iris.data[:,2].mean(),2))
print('---------------')
print('min of Petal Width: ', iris.data[:,3].min())
print('max of Petal Width: ', iris.data[:,3].max())
print('mean of Petal Width: ', round(iris.data[:,3].mean(),2))

```

### 2. Plot the figure of every two features like above.

```python
sl = iris.data[:,0]
sw = iris.data[:,1]
pl = iris.data[:,2]
pw = iris.data[:,3]
y = iris.target

fig = plt.figure(2, figsize=(16, 14))

# sepal length & sepal width
p1 = fig.add_subplot(3,2,1)
p1.scatter(sl, sw, c=y)
plt.xlabel('sepal length')
plt.ylabel('sepal width')

# petal length & sepal width
plt.subplot(3,2,2)
plt.scatter(pl, sw, c=y)
plt.xlabel('petal length')
plt.ylabel('sepal width')

# sepal length & petal width
plt.subplot(3,2,3)
plt.scatter(sl, pw, c=y)
plt.xlabel('sepal length')
plt.ylabel('petal width')

# sepal length & petal length
p1 = fig.add_subplot(3,2,1)
p1.scatter(sl, pl, c=y)
plt.xlabel('sepal length')
plt.ylabel('petal length')

# sepal width & petal width
plt.subplot(3,2,4)
plt.scatter(sw, pw, c=y)
plt.xlabel('sepal width')
plt.ylabel('petal width')


plt.show()
```

### 3. Train a Naive Bayes Classifier on the Iris dataset. (Reference: http://scikit-learn.org/stable/modules/naive_bayes.html)

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(iris.data, iris.target)
model
```

### 4. Use your classifier to answer: which types do you think these flowers belong to?
(1) a flower with Seqal (length=4.7, width=3.2), Pedal (length=1.6, width=0.2) (is it same as you think in question 1？)

(2) a flower with Seqal (length=5.2, width=2.7), Pedal (length=3.9, width=1.4)

(3) a flower with Seqal (length=6, width=2.2), Pedal (length=5, width=1.5)

```python
model.predict([[4.7, 3.2, 1.6, 0.2],
              [5.2, 2.7, 3.9, 1.4],
              [6, 2.2, 5, 1.5]])
```
